```python
import urllib.request
from bs4 import BeautifulSoup as BS
```

```python
def make_dict(user):
    url = 'https://letterboxd.com/' + user + '/films/page/1/'
    fp = urllib.request.urlopen(url)

    mybytes = fp.read()
    mystr = mybytes.decode("utf8")

    fp.close()

    soup = BS(mystr, "html.parser")

    pages = soup.find_all('li', attrs = {'class' : 'paginate-page'})

    for page in pages:
        total_pages = page.text

    movies_titles = []
    movie_ratings = []

    for i in range(1, int(total_pages) + 1):
        url = 'https://letterboxd.com/' + user + '/films/page/' + str(i) + '/'

        fp = urllib.request.urlopen(url)

        mybytes = fp.read()
        mystr = mybytes.decode("utf8")

        fp.close()

        soup = BS(mystr, "html.parser")

        divs = soup.find_all('li', attrs = {'class' : 'poster-container'})

        for i in range(len(divs)):
            img_label = divs[i].find_all('img')
            movie_name = img_label[0].attrs['alt']

            movies_titles.append(movie_name)

            rating_label = divs[i].find_all('span', attrs = {'class' : 'rating'})
            if rating_label != []:
                rating_text = rating_label[0].attrs['class'][-1]
                movie_rating = int(''.join(filter(str.isdigit, rating_text)))
            else:
                movie_rating = 'NA'

            movie_ratings.append(movie_rating)

    movies_dict = {}

    for i in range(len(movies_titles)):
        movies_dict[movies_titles[i]] = movie_ratings[i]

    return movies_dict
```

```python
user_one = 'oliviadc'
user_two = 'elfermins'
```

```python
movies_dict_user_one = make_dict(user_one)
movies_dict_user_two = make_dict(user_two)
```

```python
movies_dict_user_one_only_rated = {key:val for key, val in movies_dict_user_one.items() if val != 'NA'}
movies_dict_user_two_only_rated = {key:val for key, val in movies_dict_user_two.items() if val != 'NA'}
```

```python
# check "compatibility"
# of shared movies, compute difference

shared_movies = []

for movie in movies_dict_user_one_only_rated:
    if movie in movies_dict_user_two_only_rated:
        shared_movies.append(movie)

print('Shared movies:', len(shared_movies))
total_sum = 0

for movie in shared_movies:
    rating_from_user_one = movies_dict_user_one_only_rated[movie]
    rating_from_user_two = movies_dict_user_two_only_rated[movie]

    total_sum += abs(rating_from_user_one - rating_from_user_two)

score = 10 - total_sum / len(shared_movies)

final_score = str(int(score * 10))

print(final_score + '%', 'compatibility')
```

```python
best_ratings = [9, 10]

movies_dict_user_one_top_movies = {key:val for key, val in movies_dict_user_one.items() if val in best_ratings}
movies_dict_user_two_top_movies = {key:val for key, val in movies_dict_user_two.items() if val in best_ratings}
```
